```python
df = pd.read_excel('AdventureWorks.xlsx')
```

```python
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
# Qual a receita total ?
df['Valor Venda'].sum()
```

```python
# Qual o custo total ?
df['Custo'] = df['Custo Unitário'].mul(df['Quantidade'])
```

```python
df.head(5)
```

```python
# Qual o custo total ?
round(df['Custo'].sum(),2)
```

```python
# Encontrar o lucro total
# Lucro = Receita - custo
df['Lucro'] = df['Valor Venda'] - df['Custo']
```

```python
df.head(5)
```

```python
# Lucro total
round(df['Lucro'].sum(),2)
```

```python
# Criar um coluna com total de dias usados para enviar o produto
# Diferenã de datas
df['Tempo Envio'] = df['Data Envio'] - df['Data Venda']
```

```python
df.head(5)
```

```python
# Agora, deseja-se saber a média do tempo de envio para cada marca
# Primeiramente, é necessário transforma Tempo envio em númerico
df['Tempo Envio'] = (df['Data Envio'] - df['Data Venda']).dt.days
```

```python
df.head(5)
```

```python
df['Tempo Envio'].dtypes
```

```python
# Média do tempo de envio por Marca - Agrupar por marca
df.groupby('Marca')['Tempo Envio'].mean()
```

# Valores ausentes

```python
# Verificar se há valores ausentes
df.isnull().sum()
```

```python
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
# Lucro por ano e por marca
df.groupby([df['Data Venda'].dt.year, 'Marca'])['Lucro'].sum()
```

```python
# Resetando o index
lucro_ano = df.groupby([df['Data Venda'].dt.year, 'Marca'])['Lucro'].sum().reset_index()
lucro_ano
```

```python
# Qual o total de produtos vendidos?
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending= False)
```

```python
# Gráfico Total de produtos vendidos
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending= True).plot.barh(title= 'Total de Produtos Vendidos')
plt.xlabel('Total')
plt.ylabel('Produto');
```

```python
# Lucro por ano
df.groupby(df['Data Venda'].dt.year)['Lucro'].sum().plot.bar(title= 'Lucro x Ano')
plt.xlabel('Ano')
plt.ylabel('Receita');
```

```python
df.groupby(df['Data Venda'].dt.year)['Lucro'].sum()
```

```python
# Selecionando as vendas de 2009
df_2009 = df[df['Data Venda'].dt.year == 2009]
```

```python
df_2009.head(5)
```

```python
df_2009.groupby(df_2009['Data Venda'].dt.month)['Lucro'].sum().plot(title= 'Lucro x Mês')
plt.xlabel('Mês')
plt.ylabel('Lucro');
```

```python
df_2009.groupby(df_2009['Marca'])['Lucro'].sum().plot.bar(title= 'Lucro x Marca')
plt.xlabel('Marca')
plt.ylabel('Lucro')
plt.xticks(rotation= 'horizontal');
```

```python
df_2009.groupby(df_2009['Classe'])['Lucro'].sum().plot.bar(title= 'Lucro x Classe')
plt.xlabel('Classe')
plt.ylabel('Lucro')
plt.xticks(rotation= 'horizontal');
```

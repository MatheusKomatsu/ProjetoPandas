```python
# Renomeando o nome das colunas para a tradução em português
df = df.rename(columns= {"country": "País", "continent": "Continente", "year": "Ano", "lifeExp": "Expectativa de vida", "pop": "População total", "gdpPercap":"PIB"})
df.head(10)
```

```python
# Retornar apenas os valores únicos em continentes
df['Continente'].unique()
```

```python
# Localiza as colunas cujo continente é a oceania
Oceania = df.loc[df['Continente'] == 'Oceania']
Oceania.head()
```

```python
Oceania['Continente'].unique()
```

```python
# Aplicando agrupamento
# Agrupa por continente, e conta a quantidade de países com nome único
df.groupby('Continente')['País'].nunique()
```

```python
# Calcula a média da expectativa de vida em cada ano
df.groupby('Ano')['Expectativa de vida'].mean()
```

```python
# Calcula a média do PIB de cada continente
df.groupby('Continente')['PIB'].mean()
```

```python
# Calcula a soma do PIB de cada continente
df.groupby('Continente')['PIB'].sum()
```
